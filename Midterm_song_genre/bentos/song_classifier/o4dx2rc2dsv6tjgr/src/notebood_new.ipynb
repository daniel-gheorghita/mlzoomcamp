# Predict song genre based on audio features (Spotify dataset)

```python
from pathlib import Path
import os
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif
from sklearn.feature_extraction import DictVectorizer
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import mutual_info_score, mean_squared_error, accuracy_score, recall_score, f1_score, roc_auc_score, roc_curve
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
import xgboost as xgb 
```

# Load the raw dataset

Download from: https://www.kaggle.com/datasets/siropo/spotify-multigenre-playlists-data

The dataset is already split into multiple files based on its Spotify category: 
- rock
- blues
- metal
- hiphop
- indie
- pop
- indie

A single Pandas DataFrame will be created by concatenating all the files and adding an extra column that records the category. 

The resulting DataFrame will be saved as a new CSV file ("spotify_songs.csv").

```python
input_folder = './raw_data' 
output_folder = "./data"

# Get the files from the path provided in the OP
files = Path(input_folder).glob('*.csv')  # .glob to get subdirectories
df_list = list()
for f in files:
    print(f"Found music file {f}.")
    data = pd.read_csv(f)
    # .stem is method for pathlib objects to get the filename w/o the extension
    data['category'] = f.stem.split('_')[0] # add category column from the filename
    df_list.append(data)

df = pd.concat(df_list, ignore_index=True)
df.to_csv(os.path.join(output_folder, "spotify_songs.csv"), index=False)
df.head()
```

```python
# Extract target column
target_column = 'category'
#y = df.pop(target_column)
print(df[target_column].value_counts())
remapping_to_numbers = {"rock" : 0, 
                        "indie" : 1, 
                        "pop": 2, 
                        "metal" : 3, 
                        "hiphop": 4,
                        "alternative": 5,
                        "blues": 6}

remapping_to_numbers = {"rock" : 1, 
                        "indie" : 0, 
                        "pop": 0, 
                        "metal" : 1, 
                        "hiphop": 0,
                        "alternative": 1,
                        "blues": 0}

df[target_column]=df[target_column].replace(remapping_to_numbers)
df[target_column].value_counts()
```

```python
# Splitting the dataset 60%-20%-20% (train-val-test)
#df_train_val, df_test, y_train_val, y_test = train_test_split(df, y, test_size = 0.2, random_state=1)
#df_train, df_val, y_train, y_val = train_test_split(df_train_val, y_train_val, test_size = 0.25, random_state=1)

df_train_val, df_test = train_test_split(df, test_size = 0.2, random_state=1)
df_train, df_val = train_test_split(df_train_val, test_size = 0.25, random_state=1)
```

# Basic feature visualization

```python
# Histograms
columns_with_long_tails = ['loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'duration_ms']
for column in numeric_columns:
    if column in columns_with_long_tails:
        if column == 'loudness': # loudness is a special case because it has negative values
            sns.histplot(np.log1p(df_train[column] + abs(df_train[column].min())))
        else:
            sns.histplot(np.log1p(df_train[column]))
    else:
        sns.histplot(df_train[column])
    plt.show()
    
for column in categorical_columns:
    print(df_train[column].value_counts())
    sns.histplot(df_train[column])
    plt.show()
    
#sns.histplot(df_train[target_column])
#plt.show()
```

# Initial feature selection (based on feature distribution)

```python
# Select columns based on their distributions (nice spread)
selected_columns = ['danceability', 
                    'energy', 
                    'loudness', # borderline nice-looking
                    'liveness', 
                    'valence', 
                    'tempo', 
#                    'duration_ms', 
#                    'popularity',
                    target_column]

# Keep only the selected columns
df_train = df_train[selected_columns]
df_val = df_val[selected_columns]
df_test = df_test[selected_columns]

# Apply the log1p to the columns that were displayed with log1p and are selected
for column in selected_columns:
    if column in columns_with_long_tails:
        if column == 'loudness': # loudness is a special case because it contains negative values
            shifting_value = abs(df_train[column].min())
            df_train[column] = np.log1p(df_train[column] + shifting_value)
            df_val[column] = np.log1p(df_val[column] + shifting_value)
            df_test[column] = np.log1p(df_test[column] + shifting_value)
        else:
            df_train[column] = np.log1p(df_train[column])
            df_val[column] = np.log1p(df_val[column])
            df_test[column] = np.log1p(df_test[column])
            
# Update numeric and categorical columns lists
numeric_columns = [col for col in numeric_columns if col in selected_columns]
categorical_columns = [col for col in categorical_columns if col in selected_columns]
```

```python
# Display pairplot features and target value
sns.pairplot(df_train, hue="category")
plt.show()
```

# Feature importance

```python
# Using ANOVA-f test
feature_columns = [col for col in df_train.columns.values if col != target_column]

# configure to select all features
feature_selection_annova_f = SelectKBest(score_func=f_classif, k='all')
# learn relationship from training data
feature_selection_annova_f.fit(df_train[feature_columns].values, df_train[target_column].values)
for i in range(len(feature_selection_annova_f.scores_)):
    print(f"Feature {feature_columns[i]}: {feature_selection_annova_f.scores_[i]}")
```

```python
# Select features
selected_columns = ['danceability',
                    'energy',
                    target_column]

df_train = df_train[selected_columns]
df_val = df_val[selected_columns]
df_test = df_test[selected_columns]
```

# Extract target value and features

```python
y_train = df_train.pop(target_column).values
X_train = df_train.values
```

```python
y_val = df_val.pop(target_column).values
X_val = df_val.values
```

```python
y_test = df_test.pop(target_column).values
X_test = df_test.values
```

# Logistic Regression

```python
# Train
log_regressor = LogisticRegression()
log_regressor.fit(X_train, y_train)
```

```python
# Validation
y_pred_val = log_regressor.predict(X_val)
y_proba_val = log_regressor.predict_proba(X_val)[:,1]
print("accuracy_score: ", accuracy_score(y_val, y_pred_val))
print("recall_score: ", recall_score(y_val, y_pred_val))
print("f1_score: ", f1_score(y_val, y_pred_val))
print("roc_auc_score: ", roc_auc_score(y_val, y_proba_val))
```

# Decision Tree Classifier

```python
max_depth_list = [1, 3, 5, 10, 50, 100]
min_samples_leaf_list = [1, 5, 10, 50, 100]

dt_roc_auc_scores = np.zeros((len(max_depth_list), len(min_samples_leaf_list)))

for md_idx, md in enumerate(max_depth_list):
    for msl_idx, msl in enumerate(min_samples_leaf_list):
        # Train
        dt_classifier = DecisionTreeClassifier(max_depth=md, min_samples_leaf=msl, random_state=1)
        dt_classifier.fit(X_train, y_train)
        
        # Validation
        y_pred_val = dt_classifier.predict(X_val)
        y_proba_val = dt_classifier.predict_proba(X_val)[:,1]
        #print(accuracy_score(y_val, y_pred_val))
        #print(recall_score(y_val, y_pred_val, average=None))
        #print(f1_score(y_val, y_pred_val, average=None))
        #print(f"max_depth: {md}, min_samples_leaf: {msl}, roc_auc_score: {roc_auc_score(y_val, y_proba_val)}")
        dt_roc_auc_scores[md_idx, msl_idx] = roc_auc_score(y_val, y_proba_val)
        
sns.heatmap(dt_roc_auc_scores, annot=True, yticklabels=max_depth_list, xticklabels=min_samples_leaf_list)
plt.title("ROC AUC Score on validation set for Decision Tree Classifiers")
plt.xlabel('min_samples_leaf')
plt.ylabel('max_depth')
plt.show()
```

# Random Forest Classifier

```python
max_depth_list = [1, 3, 5, 10, 50, 100]
n_estimators_list = [10, 50, 100, 500]
min_samples_leaf_list = [1, 5, 10, 50, 100]

for ne_idx, ne in enumerate(n_estimators_list):
    rf_roc_auc_scores = np.zeros((len(max_depth_list), len(min_samples_leaf_list)))
    for md_idx, md in enumerate(max_depth_list):
        for msl_idx, msl in enumerate(min_samples_leaf_list):
            # Train
            rf_classifier = RandomForestClassifier(n_estimators=ne, max_depth=md, min_samples_leaf=msl, random_state=1)
            rf_classifier.fit(X_train, y_train)        

            # Validation
            y_pred_val = rf_classifier.predict(X_val)
            y_proba_val = rf_classifier.predict_proba(X_val)[:,1]
            #print(accuracy_score(y_val, y_pred_val))
            #print(recall_score(y_val, y_pred_val, average=None))
            #print(f1_score(y_val, y_pred_val, average=None))
            #print(f"max_depth: {md}, min_samples_leaf: {msl}, roc_auc_score: {roc_auc_score(y_val, y_proba_val)}")
            rf_roc_auc_scores[md_idx, msl_idx] = roc_auc_score(y_val, y_proba_val)

    sns.heatmap(rf_roc_auc_scores, annot=True, yticklabels=max_depth_list, xticklabels=n_estimators_list)
    plt.title(f"ROC AUC Score on validation set for Random Forrest Classifiers with {ne} estimators")
    plt.xlabel('min_samples_leaf')
    plt.ylabel('max_depth')
    plt.show()
```

# XGB Classifier

```python
# Create DMatrices
dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=df_train.columns.values)
dval = xgb.DMatrix(X_val, label=y_val, feature_names=df_train.columns.values)
dtest = xgb.DMatrix(X_test, label=y_test, feature_names=df_train.columns.values)
```

```python
eta_list = [0.1, 0.3, 0.5, 1.0]
n_estimators_list = [10, 50, 100, 500]
min_samples_leaf_list = [1, 5, 10, 50, 100]

for e_idx, e in enumerate(eta_list):
    xgb_roc_auc_scores = np.zeros((len(max_depth_list), len(min_samples_leaf_list)))
    for md_idx, md in enumerate(max_depth_list):
        for msl_idx, msl in enumerate(min_samples_leaf_list):
            
            # Training
            watchlist = [(dtrain,'train'), (dval,'eval')]
            watchlist = None

            xgb_params = {
                'eta': e, 
                'max_depth': md,
                'min_child_weight': msl,
                'objective': 'reg:logistic',
                'eval_metric': 'auc',

                'nthread': 8,
                'seed': 1,
                'verbosity': 1,
            }

            xgb_classifier = xgb.train(xgb_params, dtrain, evals=watchlist, num_boost_round=50)
            
            # Validation
            y_pred_val = xgb_classifier.predict(dval)
            xgb_roc_auc_scores[md_idx, msl_idx] = roc_auc_score(y_val, y_pred_val)
    
    sns.heatmap(xgb_roc_auc_scores, annot=True, yticklabels=max_depth_list, xticklabels=n_estimators_list)
    plt.title(f"ROC AUC Score on validation set for XGBoost Classifiers with eta={e}")
    plt.xlabel('min_samples_leaf')
    plt.ylabel('max_depth')
    plt.show()
```

# Train final model

```python
# Select XGB Classifier model based on the previous experiments
xgb_params = {
                'eta': 0.1, 
                'max_depth': 3,
                'min_child_weight': 10,
                'objective': 'reg:logistic',
                'eval_metric': 'auc',

                'nthread': 8,
                'seed': 1,
                'verbosity': 1,
            }


# Retrain on training and validation sets
X_train_full = np.vstack((X_train, X_val))
y_train_full = np.concatenate((y_train, y_val))
dtrain_full = xgb.DMatrix(X_train_full, label=y_train_full, feature_names=df_train.columns.values)
watchlist = [(dtrain_full,'train'), (dtest,'eval')]
xgb_classifier = xgb.train(xgb_params, dtrain_full, evals=watchlist, num_boost_round=50)
```

```python
# Evaluation on the training set
y_pred_train_full = xgb_classifier.predict(dtrain_full)
fpr_train_full, tpr_train_full, thresholds_test = roc_curve(y_train_full, y_pred_train_full)
print(f"Training roc_auc_score: {roc_auc_score(y_train_full, y_pred_train_full)}")

# Evaluation on  the test set
y_pred_test = xgb_classifier.predict(dtest)
fpr_test, tpr_test, thresholds_test = roc_curve(y_test, y_pred_test)
print(f"Test roc_auc_score: {roc_auc_score(y_test, y_pred_test)}")

# Plotting ROCs
plt.plot(fpr_train_full, tpr_train_full, color='blue', label="final model on full training set")
plt.plot(fpr_test, tpr_test, color='green', label="final model on test set")

plt.plot([0, 1], [0, 1], color='black', label="random model")
plt.plot([0,0,1], [0,1,1], color='red', label="ideal model")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC for the final selected model")
plt.legend()
plt.show()
```
